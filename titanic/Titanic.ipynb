```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv(r"C:\Users\Santanu Das\Desktop\titanic\train.csv")
test = pd.read_csv(r"C:\Users\Santanu Das\Desktop\titanic\test.csv")
```

```python
train.head()
```

```python
df = train.drop(['Name', 'Ticket', 'Cabin'], axis=1)
df.head()
```

```python
df.Sex[df.Sex == 'male'] = 1
df.Sex[df.Sex == 'female'] = 0
df.head()
```

```python
df1 = df.fillna(df[['Age']].mean(axis = 0))
df1
```

```python
pd.unique(df1.Embarked)
```

```python
df1.groupby('Embarked').count()
```

```python
df1.head()
```

```python
df1 = df1.assign(Embarked=(df1['Embarked'] == 'S').astype(int))
```

```python
df1.head()
```

```python
X = df1.drop(['PassengerId', 'Survived'], axis=1)
y = df1['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(iterations=1000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn import metrics 
metrics.accuracy_score(y_test, y_pred)*100
```
